## Feature Engineering

### A. Income Band

```python
df['IncomeBand'] = pd.cut(
    df['MonthlyIncome'],
    bins=[0, 3000, 6000, 9000, 12000, df['MonthlyIncome'].max()],
    labels=['Very Low', 'Low', 'Medium', 'High', 'Very High']
)
```

### B. Age Groups

```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[17, 25, 35, 45, 55, 100],
    labels=['18-25', '26-35', '36-45', '46-55', '55+']
)
```

### C. Total Satisfaction Level

```python
satisfaction_cols = [
    'EnvironmentSatisfaction',
    'JobSatisfaction',
    'WorkLifeBalance'
]

df['TotalSarisfaction'] = df[satisfaction_cols].mean(axis=1)
```

### D. Tenure Group

```python
df['TenureGroup'] = pd.cut(
    df['YearsAtCompany'],
    bins=[-1, 1, 3, 7, 15, df['YearsAtCompany'].max()],
    labels=['0-1 yrs', '2-3 yrs', '4-7 yrs', '8-15 yrs', '15+ yrs']
)
```

### E. Attrition Flag

```python
df['AttritionFlag'] = df['Attrition'].map({'Yes':1, 'No':0})
```

### F. Experience Level

```python
df['ExperienceLevel'] = pd.cut(
    df['NumCompaniesWorked'],
    bins=[-1, 1, 3, 6, df['NumCompaniesWorked'].max()],
    labels = ['Low', 'Medium', 'High', 'Very High']
)
```
